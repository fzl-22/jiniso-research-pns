```python
import pandas as pd
import numpy as np
import folium
from geopy.distance import geodesic
```

```python
df_3second = pd.read_csv("./locations/3second-analysis-latlng.csv")
df_erigo = pd.read_csv("./locations/erigo-analysis-latlng.csv")
df_uniqlo = pd.read_csv("./locations/uniqlo-analysis-latlng.csv")
```

```python
# Merge the dataframes
df = pd.concat([df_3second, df_erigo, df_uniqlo], ignore_index=True)
df
```

```python
# Create a map centered around an average latitude and longitude
average_lat = df['latitude'].mean()
average_lon = df['longitude'].mean()

# Create a folium map centered on the average latitude and longitude
m = folium.Map(location=[average_lat, average_lon], zoom_start=5)
```

```python
for index, row in df.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    brand = row['brand']
    
    # Define color based on brand
    if brand == 'Uniqlo':
        color = 'red'
    elif brand == 'Erigo':
        color = 'green'
    elif brand == '3second':
        color = 'blue'
    else:
        color = 'gray'  # Default color for any other brands
    
    # Add marker to the map
    folium.Marker(
        location=[latitude, longitude],
        popup=f"{row['name']} - {row['address']}",
        icon=folium.Icon(color=color)
    ).add_to(m)
```

```python
def find_clusters(df, radius_km=5):
    clusters = []
    for i, row in df.iterrows():
        store_location = (row['latitude'], row['longitude'])
        nearby_stores = [store_location]
        
        for j, other_row in df.iterrows():
            if i != j:
                other_store_location = (other_row['latitude'], other_row['longitude'])
                distance = geodesic(store_location, other_store_location).km
                if distance <= radius_km:
                    nearby_stores.append(other_store_location)
        
        if len(nearby_stores) >= 5:  # More than 5 stores within 5km
            clusters.append(store_location)
    
    return clusters

clusters = find_clusters(df)
```

```python
for cluster in clusters:
    folium.Circle(
        location=cluster,
        radius=5000,  # 5 km radius
        color='purple',
        fill=True,
        fill_opacity=0.1
    ).add_to(m)
```

```python
m.save('./locations/stores_map_with_clusters.html')
```
