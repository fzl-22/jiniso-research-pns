# 3second

```python
import requests
import pandas as pd
```

```python
URL = "https://3second.co.id/api/product/store"
response = requests.get(URL)
response.raise_for_status() 
```

```python
data = response.json()

stores = data['data']
parsed_data = [
    {
        "brand": "3second",
        "name": store['name'],
        "address": store['address'],
        "latitude": store['latitude'],
        "longitude": store['longitude'],
    }
    for store in stores
]

df = pd.DataFrame(parsed_data)
df
```

```python
df.to_csv("./locations/3second-analysis-latlng.csv", index=False)
```

# Erigo

```python
import requests
import pandas as pd
import json
import re
```

```python
file_path = './erigo-store.json'

with open(file_path, 'r') as file:
    data = json.load(file)

df = pd.json_normalize(data)

df
```

```python
def extract_lat_lng(iframe_code):
    # Adjust pattern to capture longitude after !2d and latitude after !3d
    pattern = r'!2d(-?\d+\.\d+)!3d(-?\d+\.\d+)'
    match = re.search(pattern, iframe_code)
    
    if match:
        longitude = float(match.group(1))
        latitude = float(match.group(2))
        return latitude, longitude
    else:
        return None, None  # If no match, return None
```

```python
# Apply the function to extract latitude and longitude
df[['latitude', 'longitude']] = df['google_maps_embed'].apply(
    lambda x: pd.Series(extract_lat_lng(x))
)
```

```python
df_transformed = pd.DataFrame({
    'brand': 'Erigo',  # Set constant brand value
    'name': df['name'],
    'address': df['address'],
    'latitude': df['latitude'],
    'longitude': df['longitude']
})

df_transformed
```

```python
df_transformed.to_csv("./locations/erigo-analysis-latlng.csv", index=False)
```

#  Uniqlo

```python
import pandas as pd
import requests
```

```python
url = "https://map.uniqlo.com/id/api/storelocator/v1/id/stores?limit=100&RESET=true&lang=local&offset=0&r=storelocator"
response = requests.get(url)
data = response.json()
```

```python
stores = data['result']['stores']
df = pd.DataFrame(stores)
df['brand'] = 'Uniqlo'
df
```

```python
df = df[['brand', 'name', 'address', 'latitude', 'longitude']]
df
```

```python
df.to_csv("./locations/uniqlo-analysis-latlng.csv", index=False)
```
